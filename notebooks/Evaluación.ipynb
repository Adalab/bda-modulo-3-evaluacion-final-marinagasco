```python
#Función que explora los datos: 

#Función que explora los datos: 

def exploracion(dataframe):
    '''
    Realiza una exploración inicial de un DataFrame de pandas proporcionando información 
    sobre los datos nulos, duplicados, tipos de datos, valores únicos y estadísticas descriptivas.

    Esta función imprime un resumen con:
    - El número de filas y columnas del DataFrame.
    - El porcentaje de valores nulos y no nulos por columna.
    - El tipo de dato de cada columna.
    - El número de valores únicos en cada columna.
    - El número total de registros duplicados y su porcentaje respecto al total de filas.
    - Las columnas que contienen datos nulos y las que no contienen datos nulos.
    - Estadísticas descriptivas para las columnas numéricas y categóricas (de tipo 'object').

    Parámetros:
    ----------
    df : pandas.DataFrame
        El DataFrame que se desea explorar.

    Retorna:
    --------
    df_info : pandas.DataFrame
        Un DataFrame con la siguiente información para cada columna:
        - '% nulos': Porcentaje de valores nulos.
        - '% no_nulos': Porcentaje de valores no nulos.
        - 'tipo_dato': Tipo de dato de la columna.
        - 'num_valores_unicos': Número de valores únicos en la columna.'''
    
    df_info = pd.DataFrame()

    df_info["% nulos"] = round(dataframe.isna().sum()/dataframe.shape[0]*100, 2).astype(str)+"%" # type: ignore
    df_info["% no_nulos"] = round(dataframe.notna().sum()/dataframe.shape[0]*100, 2).astype(str)+"%" # type: ignore
    df_info["tipo_dato"] = dataframe.dtypes # type: ignore
    df_info["num_valores_unicos"] = dataframe.nunique() # type: ignore

    print(f"""El DataFrame tiene {dataframe.shape[0]} filas y {dataframe.shape[1]} columnas. # type: ignore # type: ignore # type: ignore # type: ignore # type: ignore # type: ignore # type: ignore # type: ignore # type: ignore # type: ignore
    Tiene {dataframe.duplicated().sum()} datos duplicados, lo que supone un porcentaje de {round(dataframe.duplicated().sum()/dataframe.shape[0], 2)}% de los datos.

    Hay {len(list(df_info[(df_info["% nulos"] != "0.0%")].index))} columnas con datos nulos, y son: 
    {list(df_info[(df_info["% nulos"] != "0.0%")].index)}

    y sin nulos hay {len(list(df_info[(df_info["% nulos"] == "0.0%")].index))} columnas y son: 
    {list(df_info[(df_info["% nulos"] == "0.0%")].index)}

    A continuación tenemos un detalle sobre los datos nulos y los tipos y número de datos:""")

    display(df_info.head())

    print("Principales estadísticos de las columnas categóricas:")

    display(dataframe.describe(include="o").T)

    print("Principales estadísticos de las columnas numéricas:")

    display(dataframe.describe(exclude="o").T)

    return df_info
```

### PASO 4: 

#### Limpiamos los datos tras haber analizado nuestros dataframes: 

```python
#Empezamos por la columna "Salary" que es la primera que queremos rellenar nulos. 
#Cómo no tenemos muy claro si sería mejor hacer una media o una mediana (dependiendo de si los datos son muy extremos), vamos a visualizar un histograma que nos indicará la distribución (frecuencia) de 1 variable. 

sns.histplot(x="Salary", data=df_final, bins = 20, kde=True, color = "violet")

# usando 'plt.xlabel()' cambiamos el nombre del eje x
plt.xlabel("Salario")

# usando el método 'plt.ylabel()' cambiamos el nombre del eje y
plt.ylabel("Empleados");

# ponemos título a la gráfica
plt.title("Salario de los empleados")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;

#A la gráfica le añadimos el 'kde' que es una línea que nos muestra el patrón que siguen nuestros datos. 
#También añadimos los "bins"que es el número de barras en nuestra visualización que le indicamos. 

#A simple vista, podemos ver que la distribución de Salary es asimétrica, con una mayoría de valores concentrados en torno a los 50,000-100,000, pero con valores extremos que llegan a 400,000 o más.
#Esto indica la presencia de valores atípicos (outliers) que pueden distorsionar la media.
```

```python
#Para poder visualizar estos outliers de una forma más gráfica, hacemos un boxplot que muestra de igual forma la distribución pero incluye la mediana y etoslos valores atípicos. 
#Lo que está entre los bigotitos es lo típico en nuestros datos, lo habitual. 

sns.boxplot(x="Salary", data=df_final)

# cambiamos los nombres de los ejes
plt.xlabel("Salario")
plt.ylabel("Emnpleados")

# ponemos título a la gráfica
plt.title("Salario de los empleados")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;

#Lo que nos confirma este boxplot es que hay presencia de muchos valores atípicos (outliers).
#Esto significa que la media será influenciada por estos outliers, resultando en un valor más alto que no representa el centro de la mayoría de los datos.
#Por lo tanto, llegamos a la conclusión que debemos tomar la mediana para rellenar estos nulos de la columna "Salary". 
```

```python
# La pirmera columna con más nulos es salary. 
# Imputaremos estos nulos de "Salary" con la mediana (promedio). Rellenamos estos datos con el método "fillna". 
# Añadiremos el "inplace=Treue" para sobreescirbir esos valores (también podríamos haber creado una variable nueva y guardarlos ahí).

df_final['Salary'].fillna(df_final['Salary'].median(), inplace=True)
```

```python
## ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

## Usaremos un gráfico de barras con Seaborn para visualizar los vuelos reservados agrupados por mes.

import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar los vuelos reservados por mes
vuelos_mes = df_final.groupby('Month')['Flights Booked'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(data=vuelos_mes, x='Month', y='Flights Booked', palette='viridis')
plt.title('Cantidad de vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Vuelos Reservados')
plt.xticks(range(0, 12), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.show()
```

```python
## 2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los clientes?

## Para esto, haremos un gráfico de dispersión con sns.scatterplot.

# Crear el gráfico de dispersión

sns.scatterplot(data=df_final, x='Distance', y='Points Accumulated', alpha=0.5, color='blue')
plt.title('Relación entre Distancia y Puntos Acumulados')
plt.xlabel('Distancia (millas)')
plt.ylabel('Puntos Acumulados')
plt.show()
```

```python
## 3. ¿Cuál es la distribución de los clientes por provincia o estado?

#Aquí usamos un gráfico de barras que cuenta los clientes (Province).

# Crear un gráfico de barras para contar clientes por provincia
sns.countplot(data=df_final, y='Province', order=df_final['Province'].value_counts().index, palette='coolwarm')
plt.title('Distribución de Clientes por Provincia')
plt.xlabel('Cantidad de Clientes')
plt.ylabel('Provincia')
plt.show()
```

```python
## 4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

## Un gráfico de barras que muestra el salario promedio (Salary) por nivel educativo (Education).

# Agrupar los datos por nivel educativo y calcular el salario promedio
avg_salary_by_education = df_final.groupby('Education')['Salary'].mean().reset_index()

# Crear un gráfico de barras
sns.barplot(data=avg_salary_by_education, x='Salary', y='Education', palette='mako')
plt.title('Salario Promedio por Nivel Educativo')
plt.xlabel('Salario Promedio')
plt.ylabel('Nivel Educativo')
plt.show()
```

```python
## 5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?
## Un gráfico circular para ver la proporción de clientes según el tipo de tarjeta (Loyalty Card).

# Calcular proporción de clientes por tipo de tarjeta
loyalty_card_counts = df_final['Loyalty Card'].value_counts()

# Crear un gráfico circular
loyalty_card_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='Set3')
plt.title('Proporción de Clientes por Tipo de Tarjeta')
plt.ylabel('')  # Eliminar etiqueta del eje Y
plt.show()
```
